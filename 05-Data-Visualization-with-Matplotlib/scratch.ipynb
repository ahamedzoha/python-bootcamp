```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'g')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## Object Oriented Way

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])


axes1.plot(x,y,'r')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Title')

axes2.plot(y,x,'b')
axes2.set_title('Y as a f(x)')
```

# Looping Through axes array
### Tuple-unpacking on fig,axes using plt.subplots instead of plt.figure

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#Does something to spacing
plt.tight_layout()

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes
```

# Indexing Subplots
As axes is a list, they can be used to index and plot in the number of subplots available 

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#Does something to spacing
plt.tight_layout()

axes[0].plot(x,y,'o')
axes[0].set_title('Axes[0]')

axes[1].plot(x,y,'ro')
axes[1].set_title('Axes[1]')
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,figsize=(8,4), dpi=100 )
ax[0].plot(x,y)
ax[1].plot(x,y)

plt.tight_layout()
```

# Legends

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x,x**2,'bo',label = 'X squared')
ax1.plot(x,x**3,'ro',label = 'X cubed')

# ax1.legend()
# ax1.legend(loc=(0.1,0.1)) #for manual location
ax1.legend(loc=0) #refer the location codes in the documentation
```

# Colors and Customization

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x,x**2,'bo', label = 'X squared') # RGB colors supported
ax1.plot(x,x**3,'r',lw=3, marker='o', ms=10, label = 'X cubed') # Transparency supported

ax1.legend(loc=0)
```

```python
fig = plt.figure(dpi=150)

ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x,x**2,
         color='#007CB0',
         linewidth = 3,
         marker='o',
         markerfacecolor='#06D6A0',
         markeredgewidth=0,
         markersize=10,
         label = 'X squared'
        ) # RGB colors supported

ax1.plot(x,x**3,
         color='purple',
         linewidth=3, 
         marker='o',
         markerfacecolor= 'red',
         markeredgewidth= 0,
         ms=10, 
         label = 'X cubed'
        ) # Transparency supported

ax1.legend(loc=0)
```

## X and Y limits

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])

ax1.plot(x, x**2, label='X squared')

ax1.set_xlim([0,1])
ax1.set_ylim([0,2])

ax1.legend(loc=0)
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
ax.set_xlim([0,100])
ax.set_ylim([0,200])
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y,'r')
ax2.plot(x,y,'r')

ax1.set_xlabel('X')
ax2.set_xlabel('X')

ax1.set_ylabel('Y')
ax2.set_ylabel('Y')

ax1.set_xlim([0,100])
ax1.set_ylim([0,200])

ax2.set_xlim([0,100])
ax2.set_ylim([0,200])

fig
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,z)
ax2.plot(x,y)

ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.set_xlabel('X')
ax2.set_ylabel('Y')

ax1.set_xlim([0,100])
ax1.set_ylim([0,10000])

ax2.set_xlim([20,22])
ax2.set_ylim([30,50])

ax2.set_title('zoom')

fig
```

```python
fig, axs = plt.subplots(nrows=1,ncols=2, figsize=(12,2))

axs[0].plot(x, y, color='blue', lw=3)

axs[1].plot(x, z, color='red', lw=3, ls='dashed')

axs[0].set_xlim([0,100])
axs[0].set_ylim([0,200])

axs[1].set_xlim([0,100])
axs[1].set_ylim([0,10000])

plt.tight_layout()
```
