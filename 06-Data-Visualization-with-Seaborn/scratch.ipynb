```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('dse_2017.csv')
```

```python
data.sample(20)
```

```python
data = data[data['trade'] != 0]
```

```python
data.sample(15)
```

```python
matrix = data.pivot_table(values='trade', index='date', columns='trading_code')
```

```python
matrix.sample(15)
```

```python
sns.heatmap(matrix.sample(50))
```

```python
data.sample(20)
```

```python
sns.lmplot(x='trade', y='volume', hue='trading_code',data=data.sample(50), height=10)
```

```python
sns.lmplot(x='opening_price', y='closing_price',data=data, aspect=1, height=10, palette='seismic')
```
